```python
documents=[]
for category in movie_reviews.categories():
    for fileid in movie_reviews.fileids(category):
        documents.append((movie_reviews.words(fileid),category))
documents[0:5]
```

```python
import random

random.shuffle(documents)
documents[0:5]
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

```python
from nltk.corpus import wordnet
 
def get_simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADJ
    else:
        return wordnet.NOUN
```

```python
from nltk.corpus import stopwords
import string

stops=stopwords.words('english')
punctuations=string.punctuation
stops+=list(punctuations)
stops=set(stops)
stops
```

```python
from nltk import pos_tag

def clean_review(words):
    output_words=[]
    for w in words:
        if w not in stops:
            pos=pos_tag([w]) 
            clean_word=lemmatizer.lemmatize(w,pos=get_simple_pos(pos[0][1]))
            output_words.append(clean_word.lower())
    return output_words
```

```python
documents=[(clean_review(document),category) for document,category in documents ] #8min
```

```python
documents[0]
```

```python
training_document=documents[:1500]
testing_document=documents[1500:]
```

```python
all_words=[]
for document in documents:
    all_words+=document[0]
```

```python
import nltk
```

```python
freq=nltk.FreqDist(all_words)
common=freq.most_common(3000)
features=[i[0] for i in common]
```

```python
features
```

```python
def get_feature_dict(words):
    current_features={}
    words_set=set(words)
    for w in features:
        current_features[w]=w in words_set
    return current_features
```

```python
training_data=[(get_feature_dict(doc),category) for doc,category in training_document]
testing_data=[(get_feature_dict(doc),category) for doc,category in testing_document]
```

```python
testing_data[0]
```

```python
from nltk import NaiveBayesClassifier
```

```python
clf=NaiveBayesClassifier.train(training_data)
```

```python
nltk.classify.accuracy(clf,testing_data)
```

Count Vectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
categories=[category for document,category in documents]
```

```python
x_data=[" ".join(document) for document,category in documents]
```

```python
x_data[0]
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x_data,categories,random_state=0)
```

```python
count_vec=CountVectorizer(max_features=2000)
a=count_vec.fit_transform(x_train)
a.todense()
```

```python
count_vec.get_feature_names()
```

```python
x_test_features=count_vec.fit_transform(x_test)
```

```python
x_test_features.todense()
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report
```

```python
svc=SVC()
svc.fit(a,y_train)
```

```python
y_pred=svc.predict(x_test_features)
svc.score(x_test_features,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc=RandomForestClassifier()
rfc.fit(a,y_train)
```

```python
y_pred1=rfc.predict(x_test_features)
```

```python
print(classification_report(y_test,y_pred1))
```
