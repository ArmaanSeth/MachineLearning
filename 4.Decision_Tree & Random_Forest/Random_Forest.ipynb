# Random Forest


```python
import pandas as pd
import numpy as np  
```

```python
titanic=pd.read_csv(r'C:\Users\Armaan\Downloads\Titanic dataset\titanic_train.csv')
s=titanic["Survived"]
titanic.drop(["Cabin","Name","Ticket","Survived"],axis=1,inplace=True)
titanic["Survived"]=s
titanic
```

```python
titanic.isnull().sum()
```

```python
def set_age(s,age):
    age=np.array(age)
    m=age.sum()
    s=age[age==s].shape
    s=s[0]
    m=m/s
    return s

def set_gender(gender):
    if(gender=="male"):
        return 1
    else:
        return 0

def set_embarked(embarked):
    if(embarked=='S'):
        return 2
    elif(embarked=='C'):
        return 1
    else:
        return 0
```

```python

titanic["Age"]=titanic["Age"].fillna(titanic["Age"].median())
titanic["Embarked"].fillna(titanic["Embarked"].mod,inplace=True)
titanic["Sex"]=titanic["Sex"].apply(set_gender)
titanic["Embarked"]=titanic["Embarked"].apply(set_embarked)
titanic
```

```python
titanic["Age"]=titanic["Age"].astype(float,type[32])
titanic["Fare"]=titanic["Fare"].astype(float,type[32])
titanic.isnull().sum()
```

```python
x=titanic.drop("Survived",axis=1)
y=titanic.Survived
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
y_per_dt=dt.predict(x_test)
dt.score(x_train,y_train),dt.score(x_test,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(random_state=0,max_depth=6)
rf.fit(x_train,y_train)
rf.score(x_train,y_train),rf.score(x_test,y_test)
```
