```python
def gender(g):
    g_=""
    g=g.lower()
    for i in range(len(g)):
        if g[i]=='-' or g[i]=='/' or g[i]=='?':
            g_+=" "
        else:
            g_+=g[i]
    l=list(g_.split(" "))
    k=0
    for i in l:
        if(i=="female" or i=="f" or i=="femake" or i=="woman" or i=="she" or i=="femail" or i=="queer" or i=="Androgyne" or i=="Agender"):
            k=1
            break
    return k

# 0-male 1-female
```

```python
g="female"
print(gender(g))
```
